```python
import numpy as np
import matplotlib.pyplot as plt

x = [0]
y = [0]
weight = (85, 7, 7, 1)

funx = [(lambda x, y: 0.85 * x + 0.04 * y),        
        (lambda x, y: -0.15 * x + 0.28 * y),      
        (lambda x, y: 0.20 * x - 0.26 * y),       
        (lambda x, y: 0)]

funy = [lambda x, y: -0.04 * x + 0.85 * y + 1.6,       
        lambda x, y: 0.26 * x + 0.24 * y + 0.44,       
        lambda x, y: 0.23 * x + 0.22 * y + 1.6,      
        lambda x, y: 0.016 * y]

queue = np.random.choice([0, 1, 2, 3], size=10000, p=[weight[0]/100, weight[1]/100, weight[2]/100, weight[3]/100])

for i in queue:
    x_val = x[-1]
    y_val = y[-1]
    x_new = funx[i](x_val, y_val)
    y_new = funy[i](x_val, y_val)
    x.append(x_new)
    y.append(y_new)
plt.plot(x,y,marker=".", linewidth=0 ,markersize = 0.3)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define functions and weights
a = np.array([[0.85, 0.04], [-0.15, 0.28], [0.20, -0.26], [0, 0.16]])
b = np.array([[0, 1.6], [0, 0.44], [0, 1.6], [0, 0]])
weights = np.array([0.85, 0.07, 0.07, 0.01]).cumsum()

# Generate points
num_points = 100000
x, y = np.zeros(num_points), np.zeros(num_points)
for i in range(1, num_points):
    r = np.random.rand()
    index = np.searchsorted(weights, r)
    x[i], y[i] = a[index] @ np.array([x[i-1], y[i-1]]) + b[index]

# Plot points
plt.figure(figsize=(8,8))
plt.plot(x, y, '.', markersize=1, color='green')
plt.axis('off')
plt.show()
```
