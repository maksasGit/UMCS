```python
def julia(z ,c):
    for i in range(100):
        z = z*z + c
        if abs(z) > 2: return False 
    return True

c = -0.1 + 0.65j
size = 300
x=[]
y=[]

for i in range(size):
    for j in range(size):
        xi = -1.5 + 3/size*i
        yi = -1.5 + 3/size*j
        if julia(xi+yi*1j,  c):
            x.append(xi)
            y.append(yi)

plt.figure(figsize=(10,10))
plt.scatter(x,y,c=y , s=0.4)
plt.show()
```

```python
def julia(z ,c):
    for i in range(150):
        z = z*z + c
        if abs(z) > 2: return i
    return 100

c = -0.8 + 0.156j
size = 1000

X,Y = np.meshgrid(np.linspace(-1.5,1.5,500),np.linspace(-1.5,1.5,500))
Z = X + Y*1j

color = np.frompyfunc(julia, 2, 1,)(Z,c).astype(np.float64) 


plt.figure(figsize=(10,10))
plt.contourf(X,Y, color ,200, levels=np.linspace(0,100,101), cmap="turbo")
plt.show()
```
